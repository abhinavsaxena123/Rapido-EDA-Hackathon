### Case Study: Rapido- Revolutionizing Urban Mobility

#### Case Background: 

Imagine a city where every individual can seamlessly navigate traffic, save time, and reduce
carbon footprints—all while enjoying affordable rides. Rapido has emerged as a game-changer in
urban mobility by offering bike taxis as a convenient solution to combat traffic congestion and
make daily commuting easier.

Founded with the vision of democratizing public transport, Rapido has steadily grown into a
popular platform with millions of users across India. With a focus on last-mile connectivity,
Rapido’s rides are affordable, fast, and efficient, making it the preferred choice for students,
professionals, and everyday commuters. However, much like any fast-growing service, Rapido
faces challenges in maintaining operational efficiency while ensuring a seamless customer experience.

The goal now is to gain deeper insights into customer behaviour, optimize pricing, and analyze
operational data to ensure continued growth and success in a competitive ride-hailing market.



By analyzing this dataset, participants will uncover patterns in customer behavior, identify
 opportunities to streamline operations, and recommend strategies to increase efficiency. Insight
 from this analysis will guide Rapido’s management team in decision-making processes related o
 pricing, operations, and customer satisfacton.


#### Key fields from the Rapido dataset include:

* Customer_ID: Unique identifier for each customer.
* Booking_Date: Date and time of ride booking.
* Ride_Fare: Amount charged for each ride.
* Payment_Method: Payment options used (e.g., UPI, Wallet, Cash).
* Promo_Code: Applied discount or offer code.
* Travel_Time: Time taken to complete each ride.
* Ride_Distance: Distance travelled in kilometres.
* City: City in which the ride occurred.
* Driver_ID: Unique identifier for each driver.
* Cancellation_Status: Whether a ride was cancelled. 0: Completed, 1: Cancelled

## Statistical Test

#### Checking Normality of data

```python
numerical_cols = ['Ride_Fare', 'Travel_Time', 'Ride_Distance', 
                    'Cost_per_km', 'Cost_per_minute', 'Booking_Hour']

for column in numerical_cols:
    data = df[column]
    print(f"\nChecking normality for: {column}")

    # Kolmogorov-Smirnov Test
    stat, p = kstest(data, 'norm')  # Comparing with a normal distribution
    print(f"Kolmogorov-Smirnov Test: stat={stat:.3f}, p={p:.3f}")

    # Anderson-Darling Test
    result = anderson(data, dist='norm')
    print(f"Anderson-Darling Test: stat={result.statistic:.3f}")
    print("Critical values:", result.critical_values)
    
    # Skewness and Kurtosis
    print(f"Skewness: {skew(data):.3f}, Kurtosis: {kurtosis(data):.3f}")
    
    # Visualization
    sns.histplot(data, kde=True, bins=30)
    plt.title(f"Histogram for {column}")
    plt.show()
    
    sm.qqplot(data, line='s')
    plt.title(f"Q-Q Plot for {column}")
    plt.show()
```

#### Summary of Results:
1. Kolmogorov-Smirnov Test: For all variables, p=0.000, indicating that the null hypothesis (data is normally distributed) is rejected.
2. Anderson-Darling Test: The test statistic for all variables far exceeds the critical values at all significance levels, strongly rejecting normality.
3. Skewness and Kurtosis:
* Ride_Fare, Travel_Time, Ride_Distance, and Booking_Hour have low skewness and kurtosis values close to zero, suggesting symmetry and light tails, but this does not imply normality.
* Cost_per_km and Cost_per_minute have high skewness and kurtosis, indicating extreme non-normality due to heavy tails or outliers.

#### Interpretation: 
* None of the columns are normally distributed based on the tests performed. Even though some columns have skewness and kurtosis values close to zero, the statistical tests and high Anderson-Darling statistics conclusively indicate non-normality.

#### 1.Comparing Ride Fares on Weekdays vs. Weekends
#### Mann-Whitney U Test (Non-Parametric Test): 
* Null Hypothesis (H₀): The median Ride Fare is the same for Weekdays and Weekends.
* Alternative Hypothesis (H₁): The median Ride Fare is different for Weekdays and Weekends.

```python
# Splitting data into two groups: Weekdays and Weekends
weekday_fares = df[df['Is_Weekend'] == 0]['Ride_Fare']
weekend_fares = df[df['Is_Weekend'] == 1]['Ride_Fare']

# Performing Mann-Whitney U Test
stat, p = mannwhitneyu(weekday_fares, weekend_fares, alternative='two-sided')
print(f"Mann-Whitney U Test: stat={stat:.3f}, p={p:.3f}")

if p < 0.05:
    print("Reject the null hypothesis: Significant difference between Weekdays and Weekends.")
else:
    print("Fail to reject the null hypothesis: No significant difference between Weekdays and Weekends.")
```

#### 2. Impact of Payment Methods on Ride Fares
#### Kruskal-Wallis Test (Non-Parametric)
* Null Hypothesis (H₀): The distribution of Ride Fares is the same across all Payment Methods.
* Alternative Hypothesis (H₁): The distribution of Ride Fares differs across at least one Payment Method.

```python
# Group Ride Fares by Payment Method
groups = [df[df['Payment_Method']==method]['Ride_Fare'] for method in df['Payment_Method'].unique()]

# Kruskal-Wallis Test
stat, p = kruskal(*groups)
print(f"Kruskal-Wallis Test: stat={stat:.3f}, p={p:.3f}")

if p < 0.05:
    print("Reject the null hypothesis: Significant difference between groups.")
else:
    print("Fail to reject the null hypothesis: No significant difference between groups.")
```

#### 3. Cost Efficiency: Cost per km and Cost per minute
#### Kruskal-Wallis H test
* Null Hypothesis (H₀): There is no significant difference in the cost efficiency (Cost per km or Cost per minute) across different cities or drivers.
* Alternative Hypothesis (H₁): There is a significant difference in the cost efficiency across at least one city or driver.

#### Cost per km vs. City:

```python
# Grouping data by City for Cost_per_km
city_groups_km = [df[df['City'] == city]['Cost_per_km'] for city in df['City'].unique()]

# Kruskal-Wallis Test for Cost per km
stat_km, p_km = kruskal(*city_groups_km)
print(f"Kruskal-Wallis Test for Cost per km vs. City: stat={stat_km:.3f}, p={p_km:.3f}")

if p_km < 0.05:
    print("Reject the null hypothesis: Significant difference in Cost per km across cities.")
else:
    print("Fail to reject the null hypothesis: No significant difference in Cost per km across cities.")
```

#### Cost per minute vs. City

```python
# Grouping data by City for Cost_per_minute
city_groups_minute = [df[df['City'] == city]['Cost_per_minute'] for city in df['City'].unique()]

# Kruskal-Wallis Test for Cost per minute
stat_minute, p_minute = kruskal(*city_groups_minute)
print(f"Kruskal-Wallis Test for Cost per minute vs. City: stat={stat_minute:.3f}, p={p_minute:.3f}")

if p_minute < 0.05:
    print("Reject the null hypothesis: Significant difference in Cost per minute across cities.")
else:
    print("Fail to reject the null hypothesis: No significant difference in Cost per minute across cities.")
```
